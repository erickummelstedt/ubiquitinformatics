```python
multimer_size = 4
# Set up output directory
output_dir = project_root / 'back_end' / 'data' / 'reaction_database' / f'multimer_size_{multimer_size}'
# Read the CSV files into DataFrames
ubiquitin_history = pd.read_csv(output_dir / "ubiquitin_history.csv")
reaction_history = pd.read_csv(output_dir / "reaction_history.csv")
donor_history = pd.read_csv(output_dir / "donor_history.csv")
context_history = pd.read_csv(output_dir / "context_history.csv")
```

```python
import ast

# Function to extract length of ABOC_lysines
def extract_aboc_length(cell_string):
    d = ast.literal_eval(cell_string)
    return len(d['ABOC_lysines'])

# Apply function to *every column* and create a new DataFrame
aboc_num_df = context_history.map(extract_aboc_length)

# Step 1: select the last column
last_column = aboc_num_df.iloc[:, -1]

# Step 2: find the max value
max_value = last_column.max()

# Step 3: filter rows where last column == max_value
filtered_rows = aboc_num_df[last_column == max_value]

# Step 4: The indexes you kept:
selected_indexes = filtered_rows.index

# Now pull those rows from the original df
ubiquitin_history_filtered = ubiquitin_history.loc[selected_indexes]
reaction_history_filtered = reaction_history.loc[selected_indexes]
donor_history_filtered = donor_history.loc[selected_indexes]
context_history_filtered = context_history.loc[selected_indexes]
reaction_history_filtered
```
